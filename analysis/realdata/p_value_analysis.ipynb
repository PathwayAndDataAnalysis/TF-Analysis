```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.multitest import multipletests
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
```

```python
p_value = pd.read_csv('data/pvalue_3valid_target.tsv', sep='\t')

tfs = ['MYC', 'ABCB1', 'ABL1']

for i in tfs:
    values = p_value[i].dropna()
    plt.hist(values, bins=300)
    plt.title(i)
    plt.show()
    
```

```python
p_value = pd.read_csv('data/test_p.tsv', sep='\t')
p_value_emp = pd.read_csv('data/test_p2.tsv', sep='\t')

tfs = ['MYC', 'ABCB1', 'ABL1']

for i in tfs:
    values1 = p_value[i].dropna()
    values2 = p_value_emp[i].dropna()
    plt.scatter(values1, values2, s=1)
    plt.title(i)
    plt.show()
```

```python
# Plotting the empirical distribution for different number of targets
dist_file = 'data/dist_255_9235_100000.npz'
dist = np.load(dist_file, allow_pickle=True)['distribution']

for i in range(2, 250, 100):
    plt.hist(dist[i], bins=100)
    mean = np.mean(dist[i])
    sd = np.std(dist[i])
    plt.title(f'Analytical distribution for {i} targets, mean: {mean}, and sd: {sd}')
    plt.show()
```

```python
# Plotting the analytical distribution for different number of targets
dist_file = 'data/dist_anal_255_9235_100000.npz'
dist = np.load(dist_file, allow_pickle=True)['distribution']

for i in range(2, 250, 100):
    plt.hist(dist[i], bins=100)
    mean = np.mean(dist[i])
    sd = np.std(dist[i])
    plt.title(f'Analytical distribution for {i} targets, mean: {mean}, and sd: {sd}')
    plt.show()
```
